```python
import selenium
from selenium import webdriver
import time
from time import sleep
from selenium.webdriver import ChromeOptions
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
```

```python
# Xpath与正则一齐与selenium使用
# 点击browser.execute_script('arguments[0].click',button)


# 设置浏览器
option = ChromeOptions()
option.add_experimental_option('excludeSwitches', ['enable-automation'])
browser = selenium.webdriver.Chrome()
wait = WebDriverWait(browser, 2)

# 进网址
main_url = 'https://www.hyey.cn'
browser.get(main_url)

# 登录
label_1 = 'UserName'
label_2 = 'PassWord'
label_3 = 'loginbtn'
label_4 = 'layui-layer-btn0'
account = 'ychfped'
password = 'ped123456'
input_account = browser.find_elements_by_id(label_1)
input_account[0].send_keys(account)
Password = browser.find_elements_by_id(label_2)
Password[0].send_keys(password)
button1 = browser.find_element_by_id(label_3)  # 获取点击按钮
button1.click()  # 点击登录
time.sleep(2)
button2 = browser.find_element_by_class_name(label_4)  # 获取点击按钮
button2.click()  # 点击确定
time.sleep(5)

# 搜索
searchmes_xpath = '/html/body/form/div[4]/div[3]/input[1]'
searchmes = browser.find_elements_by_xpath(searchmes_xpath)
searchmes[0].send_keys('头孢克肟胶囊')
button = browser.find_element_by_id("Button1")  # 获取点击按钮
button.click()  # 点击搜索
time.sleep(10)


def get_message(Path, path):
    for txt in path:
        Path.append(txt.text)
    return Path


i = 1
j = 1
Names = []
Company = []
Regular = []
Package = []
Middlepackage = []
Produce_Date = []
Save_Date = []
Ending_Date = []
Comman_price = []
Discount_price = []
Charge = []
while j < 7:
    print('开始第%d次循环'%j)
    print('开始抓取第%d页' % j)
    time.sleep(10)

    # 增加页码数字

    # 将页面拉到最上方
    print("将页面拉到最上方")
    time.sleep(5)
    browser.execute_script("window.scrollTo(0,0);")
    time.sleep(5)

    # 获取信息
    main_xpath = '/html/body/form/div[8]/div/div[5]/ul/li/div[3]'
    name_xpath = main_xpath + '/div[1]/p/span/span[1]'
    comapany_xpath = main_xpath + '/div[1]/p/span/span[3]'
    regular_xpath = main_xpath + '/div[1]/p/span/span[5]'
    package_xpath = main_xpath + '/div[1]/ul/li[7]/a'
    middlepackage_xpath = main_xpath + '/div[1]/ul/li[3]/a'
    produce_xpath = main_xpath + '/div[1]/ul/li[1]/span'
    ending_xpath = main_xpath + '/div[1]/ul/li[6]/a'
    save_xpath = main_xpath + '/div[2]/p[1]/span'
    comman_price_xpath = main_xpath + '/div[1]/ul/li[8]/a'
    price_xpath = main_xpath + '/div[1]/p/span/span[8]/a/span'
    charge_xpath = main_xpath + '/div[1]/ul/li[2]/span'

    # 解析为element对象
    name_nodes = browser.find_elements_by_xpath(name_xpath)
    comapany_nodes = browser.find_elements_by_xpath(comapany_xpath)
    regular_nodes = browser.find_elements_by_xpath(regular_xpath)
    package_nodes = browser.find_elements_by_xpath(package_xpath)
    middlepackage_nodes = browser.find_elements_by_xpath(middlepackage_xpath)
    produce_nodes = browser.find_elements_by_xpath(produce_xpath)
    save_notes = browser.find_elements_by_xpath(save_xpath)
    ending_notes = browser.find_elements_by_xpath(ending_xpath)
    comman_price_notes = browser.find_elements_by_xpath(comman_price_xpath)
    price_notes = browser.find_elements_by_xpath(price_xpath)
    charge_notes = browser.find_elements_by_xpath(charge_xpath)

    # 填充
    get_message(Names, name_nodes)
    get_message(Company, comapany_nodes)
    get_message(Regular, regular_nodes)
    get_message(Package, package_nodes)
    get_message(Middlepackage, middlepackage_nodes)
    get_message(Produce_Date, produce_nodes)
    get_message(Save_Date, save_notes)
    get_message(Ending_Date, ending_notes)
    get_message(Comman_price, comman_price_notes)
    get_message(Discount_price, price_notes)
    get_message(Charge, charge_notes)
    time.sleep(2)
    time.sleep(2)
    # 滚动页面滚动条到最底部
    print('将页面拉到底部')
    browser.execute_script("window.scrollTo(0,1000);")
    time.sleep(2)
    # 定位到页面页码，并生成按钮
    print('抓取第{}页结束，已抓取对象{}个'.format(i,len(Names)) )
    print('开始转换页码')
    i = i + 1
    print('开始转换到第%d页' % i)

    if i == 2:
        page_xpath = '/html/body/form/div[8]/div/div[7]/div/div[1]/span[1]/a[6]'
        page_button = browser.find_element_by_xpath(page_xpath)
        page_button.click()
    elif 2 < i < 7:
        page_xpath = '/html/body/form/div[8]/div/div[7]/div/div[1]/span[1]/a[8]'
        page_button = browser.find_element_by_xpath(page_xpath)
        page_button.click()
    else:
        pass
    j += 1

#结束提醒
print('抓取完成')
print('共抓取对象%d个' % len(Names))
browser.close()
```

```python
#写进excel表,用pandas处理
data = pd.DataFrame([Names,Company,Regular,Package,Middlepackage,Produce_Date,Save_Date,Ending_Date,
                     Comman_price,Discount_price,Charge]).T
data
```

```python
#修改列名
data.rename(columns={0:'药品名称',1:'生产公司',2:'规格',3:'包装',4:'中包',5:'生产日期',6:'库存',7:'失效日期',
                     8:'国家定价',9:'折扣价',10:'单位'},inplace=True)

data

#保存数据源
data.to_excel('./2021-08-21头孢克肟胶囊爬取数据.xlsx',encoding='gbk')
```
